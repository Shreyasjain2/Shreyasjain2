```python
import mediapipe as mp
import cv2
import numpy as np
```

```python
mp_drawings = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
#capture and process NOT SHOWING THE LANDMARKS

cap = cv2.VideoCapture(0)
holistic =  mp_holistic.Holistic(min_detection_confidence = 0.5, min_tracking_confidence =0.5)
while cap.isOpened():
    ret,frame = cap.read()
            
    img = cv2.cvtColor(frame ,cv2.COLOR_BGR2RGB)
    img.flags.writeable = False
    
    results= holistic.process(img)
    
    
    img = cv2.cvtColor(img ,cv2.COLOR_RGB2BGR)
    img.flags.writeable = True
    
    cv2.imshow("capturing you....",img)
    
    
    print(results)
    
    if cv2.waitKey(10) & 0xFF==ord('q') :
        break
        
    
    
cap.release()
cv2.destroyAllWindows()
```

```python
# Store names of the joints in one array to avoid calling again
def make_pos_dict():
    names={}
    for i,lndmark in enumerate(mp_holistic.PoseLandmark) :
        names[str(lndmark.name)] = i
    return names
```

```python
def coord_angle(result_landmarks,a,b,c,label=0) :
    
    names = make_pos_dict()
    try:

        a = [result_landmarks[names[a]].x ,result_landmarks[names[a]].y]
        b = [result_landmarks[names[b]].x ,result_landmarks[names[b]].y]
        c = [result_landmarks[names[c]].x ,result_landmarks[names[c]].y]


        radians = np.arctan2(c[1]-b[1] , c[0] -b[0]) - np.arctan2(a[1]-b[1] , a[0]-b[0])
        angle = abs(np.multiply(180/np.pi , radians))

        if angle>180 :
            angle =360-angle
        if label==0 :
            return b, int(angle)
        else :

            cv2.putText(
            img, 
            str(angle), 
            (int(width - (b[0] * width)), int(b[1] * height)), 
            cv2.FONT_HERSHEY_COMPLEX, 
            1, 
            (255, 0, 255), 
            2
             )
            return angle

    except :
        print("Invalid JOINT names")
```

```python
def get_coords(result_landmarks , a):
    names = make_pos_dict()
    
    try :
        coords= [result_landmarks[names[a]].x ,result_landmarks[names[a]].y]  
        return coords
    
    except:
        print("Wrong Joint Name")
    
```

```python
def TreePose(result_landmarks, img): 
    try :
        cords ,left_up_leg_angle = coord_angle(result_landmarks,'RIGHT_ANKLE','RIGHT_HIP','LEFT_KNEE',0)
        cords,right_up_leg_angle = coord_angle(result_landmarks,'LEFT_ANKLE','LEFT_HIP','RIGHT_KNEE',0)
        right_thumb = [result_landmarks[names['RIGHT_THUMB']].x*width ,result_landmarks[names['RIGHT_THUMB']].y*height]
        left_thumb = [result_landmarks[names['LEFT_THUMB']].x*width ,result_landmarks[names['LEFT_THUMB']].y*height]
        left_ind = [result_landmarks[names['LEFT_INDEX']].x *width,result_landmarks[names['LEFT_INDEX']].y*height]
        right_ind = [result_landmarks[names['RIGHT_INDEX']].x*width ,result_landmarks[names['RIGHT_INDEX']].y*height]
        
    
    
    
        if( (40<=left_up_leg_angle<=110) or (40<=right_up_leg_angle<=110)) :

            if(abs(right_thumb[0]- left_thumb[0]) < 10 and abs(right_thumb[1]- left_thumb[1]) < 10) :
                if(abs(right_ind[0]- left_ind[0])< 10 and abs(right_ind[1]- left_ind[1]) < 10) :
                    cv2.putText(
                        img, 
                        "TREE POSE", 
                        (int(width-150), 50), 
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, 
                        (0, 255,0), 
                        2
                         )
                    return True
            
    except:
        return False
    return False
            
```

```python
def Hasta_uttanasana(result_landmarks, img):
    try:
        
        coord ,right_body_angle = coord_angle(result_landmarks,'RIGHT_ANKLE','RIGHT_HIP','RIGHT_WRIST')
        coord ,left_body_angle = coord_angle(result_landmarks,'LEFT_ANKLE','LEFT_HIP','LEFT_WRIST')
        left_wrist = get_coords(result_landmarks , "LEFT_WRIST")
        right_wrist =  get_coords(result_landmarks , "RIGHT_WRIST")
        nose = get_coords(result_landmarks , "NOSE")
        right_elbow  = get_coords(result_landmarks , "RIGHT_ELBOW")
        left_elbow = get_coords(result_landmarks , "LEFT_ELBOW")
        right_shoulder = get_coords(result_landmarks , "RIGHT_SHOULDER")
        left_shoulder = get_coords(result_landmarks , "LEFT_SHOULDER")

        if(90<right_body_angle<175 and 90<right_body_angle<175):
            if(left_wrist[0] < nose[0] and right_wrist[0]< nose[0]):
                if(right_shoulder[1]>right_elbow[1] and left_shoulder[1]>left_elbow[1]):
                    if(right_elbow[1]<right_wrist[1] and left_elbow[1]<left_wrist[1] ) :
                        
                        cv2.putText(
                            img, 
                            "HASTA UTTANASANA", 
                            (int(width-150), 50), 
                            cv2.FONT_HERSHEY_SIMPLEX,
                            0.5, 
                            (0, 255,0), 
                            2
                             )
                        return True


    except:
        return False
    
            
    return False
    
    
```

```python
def Padhastasana(result_landmarks ,img):
    
    try :
        left_knee,angle_knee_left = coord_angle(result_landmarks ,'LEFT_ANKLE','LEFT_KNEE','LEFT_HIP',0)
        right_knee,angle_knee_right = coord_angle(result_landmarks ,'RIGHT_ANKLE','RIGHT_KNEE','RIGHT_HIP',0)
        left_toe = get_coords(result_landmarks , 'LEFT_FOOT_INDEX')
        right_toe = get_coords(result_landmarks , 'RIGHT_FOOT_INDEX')
        left_wrist = get_coords(result_landmarks , 'LEFT_WRIST')
        right_wrist = get_coords(result_landmarks , 'RIGHT_WRIST')


        if( 140<=angle_knee_right<=180 and 140<=angle_knee_left<=180):
            if(left_wrist[1]>left_knee[1] and right_wrist[1]>right_knee[1]):
                if(abs(right_toe[1]-right_wrist[1]) < 100 and abs(left_toe[1]-left_wrist[1]) <100):

                    cv2.putText(
                                img, 
                                "PADHAHASTASANA", 
                                (int(width-150), 50), 
                                cv2.FONT_HERSHEY_SIMPLEX,
                                0.5, 
                                (0, 255,0), 
                                2
                                 )
                    return True


    except:
        print("ye nahi run ho raha")
        return False
    
            
    return False             
                
    
    
```
